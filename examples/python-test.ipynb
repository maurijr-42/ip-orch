```python
from ip_orch import IPOrch
```

```python
orch = IPOrch()
orch.supported_models()
```

```python
orch.add_model("mace", "mace-mp")  # testing an available model
orch.add_model("mace", "mace-mpa")  # testing a non-available model
orch.supported_models()
```

```python
from ase.build import bulk


def calculators_test(calculator_name, ase_calculator):
    atoms = bulk("Cu", "fcc", a=3.6)
    atoms.calc = ase_calculator
    print(atoms.get_potential_energy())


orch.run(calculators_test, models=["mace-mp"])
```
